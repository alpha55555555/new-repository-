```python
def fetch_chat_completions(messages, max_retries=5, initial_wait_time=2):
    """
    尝试使用 OpenAI API 获取 chat completion。

    :param messages: 传递给 API 的消息列表。
    :param line_number: 当前正在处理的行号。
    :param max_retries: 最大重试次数。
    :param initial_wait_time: 初始等待时间（秒）。
    :return: (成功标志, 结果或错误消息)
    """
    for attempt in range(max_retries):
        try:
            # 假设此处有一个API客户端初始化和调用过程
            client = OpenAI(api_key=apikey_in_use)
            chat_completion = client.chat.completions.create(messages=messages, model="gpt-3.5-turbo")
            result = chat_completion.choices[0].message.content
            
            # result = "fake response for demonstration"  # 模拟API响应
            # write_log_to_csv(log_file_path, messages, result)  # 假设这里写入日志
            
            return True, result
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            
            wait_time = initial_wait_time * (2 ** attempt)  # 指数退避策略
            if attempt < max_retries - 1:
                print(f"Retrying in {wait_time} seconds...")
                time.sleep(wait_time)
            else:
                print("Max retries reached.")
                return False, None

def process_json_and_fetch_responses(json_file_path, start_line=0):
    """
    逐行读取JSON文件，每行包含一个JSON对象。从指定行开始，将每个对象转换为messages并获取响应。

    :param json_file_path: 包含JSON对象的文件路径。
    :param start_line: 开始处理的行号。
    """
    with open(json_file_path, 'r', newline='',encoding='utf-8-sig', errors='replace') as file:
        for line_number, line in enumerate(file, start=1):
            if line_number < start_line:
                continue  # 跳过直到到达开始行
            
            json_obj = json.loads(line.strip())
            if 'prompts' in json_obj:
                messages = json_obj['prompts']
                success, response = fetch_chat_completions(messages)
                if success:
                    print(f"Response for prompt {json_obj['prompts']}: {response}")
                    write_log_to_csv(log_file_path, messages, response)
                    inputresult("promptList\character_chat_output.json", messages, response)
                else:
                    print(f"Disconnected at line {line_number}.")
                    write_errorlog_to_csv(log_file_path, line_number, messages)
                    break  # 断线时停止处理
                
process_json_and_fetch_responses("promptList\character_chat_prompt3_2.json", start_line=0)
```
